```python
model=Sequential() #creamos un modelo layer-to-layer
```

```python
model.add(Dense(5,input_dim=2,activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
PesosI=model.get_weights()#guardamos los pesos iniciales
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy']) # establecemos el loss, y el optimizer que
#vamos usar, en este caso adam.
```

```python
#creamos un callback personalizado,
class CallbackPersonalizado(tf.keras.callbacks.Callback):
    def __init__(self, min_accuracy):
        super(CallbackPersonalizado, self).__init__() #heredamos los atributos
        self.min_accuracy = min_accuracy #guardamos el min acccuracy

    def on_epoch_end(self, epoch, logs=None): 
        accuracy = logs["accuracy"]
        if accuracy >= self.min_accuracy: #si es >= que 1, entonces
            self.model.stop_training = True # se diente el entrenamiento

callbackAccuracy = CallbackPersonalizado(min_accuracy=1.0) #establecemos el accuracy minimo

#esto lo hacemos para que para de entrenar cuando tenga un accuracy >=, al valor que digamos, en este caso 1
```
