```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
from tensorflow.keras.optimizers import SGD
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(36)
```

```python
model=Sequential()
```

```python
model.add(Dense(5,input_dim=2,activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
PesosI=model.get_weights()
```

```python
#model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#sgd = SGD(lr=0.1)
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([[0],[1],[1],[0]])
hist=model.fit(X, Y, batch_size=1, epochs=1000)
```

```python
model.evaluate(X, Y)
```

```python
model.summary()
```

```python
hist.history.keys()
```

```python
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper left')
plt.show()
```

```python
plt.plot(hist.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper left')
plt.show()
```

```python
model2=Sequential()
```

```python
model2.add(Dense(5,input_dim=2,activation='relu'))
model2.add(Dense(1, activation='relu'))
```

```python
model2.set_weights(PesosI)
```

```python
model2.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
hist2=model2.fit(X, Y, batch_size=1, epochs=1000)
```

```python
model2.evaluate(X, Y)
```

```python
plt.plot(hist2.history['loss'])
plt.title('model relu loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper left')
plt.show()
```

```python
plt.plot(hist2.history['accuracy'])
plt.title('model relu accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper left')
plt.show()
```
