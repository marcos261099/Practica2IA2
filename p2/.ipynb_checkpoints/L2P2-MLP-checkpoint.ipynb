## MLP con Keras

```python
#declaración de librerías
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
from tensorflow.keras.optimizers import SGD
import tensorflow as tf
from tensorflow.keras.callbacks import EarlyStopping
import matplotlib.pyplot as plt
import numpy as np
tf.random.set_seed(10) #declaramos una semilla
```

### Con Sigmoide

```python
model=Sequential() #creamos un modelo layer-to-layer
```

```python
model.add(Dense(5,input_dim=2,activation='sigmoid')) # 2 entradas, con un numero n(5) de neuronas en la capa oculta, y funcion de activacion sigmoide
model.add(Dense(1, activation='sigmoid')) # 1 salida, función de activación sigmoide
```

```python
PesosI=model.get_weights()#guardamos los pesos iniciales
```

```python
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy']) # establecemos el loss, y el optimizer que
#vamos usar, en este caso adam.
```

```python
#creamos un callback personalizado,
class CallbackPersonalizado(tf.keras.callbacks.Callback):
    def __init__(self, min_accuracy):
        super(CallbackPersonalizado, self).__init__() #heredamos los atributos
        self.min_accuracy = min_accuracy #guardamos el min acccuracy

    def on_epoch_end(self, epoch, logs=None): 
        accuracy = logs["accuracy"]
        if accuracy >= self.min_accuracy: #si es >= que 1, entonces
            self.model.stop_training = True # se diente el entrenamiento

callbackAccuracy = CallbackPersonalizado(min_accuracy=1.0) #establecemos el accuracy minimo

#esto lo hacemos para que para de entrenar cuando tenga un accuracy >=, al valor que digamos, en este caso 1
```

#### Referencias donde hemos visto el callback personalizado:
https://stackoverflow.com/questions/53500047/stop-training-in-keras-when-accuracy-is-already-1-0

https://stackoverflow.com/questions/59563085/how-to-stop-training-when-it-hits-a-specific-validation-accuracy

```python
#XOR
X=np.array([[0,0],[0,1],[1,0],[1,1]])
Y=np.array([[0],[1],[1],[0]])
#ponemos el callbacks, y ponemos un numero de epochs
hist=model.fit(X, Y, epochs=4000, callbacks=[callbackAccuracy])
```

```python
model.evaluate(X, Y) #evaluamos 
```

```python
model.summary() #resumen de nuestra red neuronal
```

```python
#las keys que podemos usar
hist.history.keys()
```

### Graficas

```python
#aqui podemos ver como ha ido disminuyendo el error a lo largo de los epoch
plt.plot(hist.history['loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper left') #ponemos la leyenda
plt.show()
```

```python
#aqui podemos ver como ha ido aumentando el accuracy a lo largo de los epochs
plt.plot(hist.history['accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper left') #leyenda
plt.show() # lo mostramos
```

### Modelo Con relu

```python
model2=Sequential()  #creamos un modelo layer-to-layer
```

```python
model2.add(Dense(5,input_dim=2,activation='relu'))  # 2 entradas, con un numero n(5) de neuronas en la capa oculta, y funcion de activacion relu
model2.add(Dense(1, activation='relu')) # 1 salida, función de activación relu
```

```python
model2.set_weights(PesosI) #ponemos los pesos con los que hemos realizado el anteriro modelo (los aleatorios, sin entrenar)
```

```python
model2.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
 # establecemos el loss, y el optimizer que
#vamos usar, en este caso adam.
```

```python
#si detecta que no vambia el loss, durante 50 iteraciones, entonces termina
# esto pasa algunas veces al utilizar la funcion relu, debido a:
####dying relu (explicado memoria)
estancado = EarlyStopping(monitor='loss', mode='min',patience=50)
```

```python
hist2=model2.fit(X, Y, epochs=4000,callbacks=[estancado,callbackAccuracy]) 
#ponemos el callbacks, y ponemos un numero de epochs.
#en el callback ponemos el explicada en la celda anterior 'estancado' y el 'callbackAccuracy'. este ultimo lo hemos
#explicado en el primer model (que se detenga cuando tenga un accuracy de 1.0)
```

```python
model2.evaluate(X, Y) # lo evaluamos
```

### Graficas

```python
#aqui podemos ver el loss a lo largo de los epochs
plt.plot(hist2.history['loss'])
plt.title('model relu loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['test'], loc='upper left') # leyenda
plt.show() # lo mostramos
```

```python
#aqui podemos ver el accuracy a lo largo de los epochs
plt.plot(hist2.history['accuracy'])
plt.title('model relu accuracy')
plt.ylabel('accuracy') # y label
plt.xlabel('epoch')# x label
plt.legend(['test'], loc='upper left')
plt.show() # lo mostramos
```
